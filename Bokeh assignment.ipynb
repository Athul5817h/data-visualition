```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure,output_file,show
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
import numpy as np
output_file("test.html")
p = figure(title = "sine wave")
p.xaxis.axis_label = "X"
p.yaxis.axis_label = "sine"
x = np.arange(0,10)
y = np.sin(x)
p.scatter(x,y)
show(p)
```

```python
#2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
```

```python
n = 100  
x = np.random.rand(n) * 100  
y = np.random.rand(n) * 100  
sizes = np.random.rand(n) * 20 + 10  
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], n)
```

```python
output_file("scatter.html")
p = figure(title = "randomly generated x and y values")
p.xaxis.axis_label = "X"
p.yaxis.axis_label = "Y"
p.scatter(x,y,color = colors, size = sizes)
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
```

```python

fruits = ["apple","orange","bannans","pears"]
counts = [23,45,23,30]


output_file("bar.html")
p = figure(x_range = fruits,title = "Bokeh bar chart representing the counts of different fruits using the following dataset.")
p.xaxis.axis_label = "fruits"
p.yaxis.axis_label = "counts"
p.vbar(fruits,counts)
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
```

```python
from bokeh.transform import hist
import numpy as np

data = np.random.normal(0, 1, 1000) 

hist, edges = np.histogram(data, bins=30)

p = figure(title='Histogram', x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white')

output_notebook()
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, output_file, save, show
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
from bokeh.layouts import layout
from bokeh.io import output_notebook

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap Example", x_axis_label='X-axis', y_axis_label='Y-axis',
           x_range=(0, 1), y_range=(0, 1))
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
show(p)
```
