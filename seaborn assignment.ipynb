```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

sns.relplot(x = np.random.rand(100), y = np.random.rand(100),kind = "line",label = "random")
plt.title("dataset of random numbers")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using aheatmap.
```

```python
ds = sns.load_dataset('tips')
```

```python
ds1 = ds[["tip","total_bill","size"]]
ds1
```

```python
sns.heatmap(ds1.corr(),cmap = "coolwarm",annot = True)
plt.title("correlation matrix of a dataset using a heatmap")
plt.show()
```
