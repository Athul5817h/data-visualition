```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.
```

```python
np.random.seed(30)
data = {

    'X': np.random.uniform(-10, 10, 300),

    'Y': np.random.uniform(-10, 10, 300),

    'Z': np.random.uniform(-10, 10, 300)

}

df = pd.DataFrame(data)
```

```python
df
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d( x = df.X, y = df["Y"], z = df.Z , mode = "markers" , marker_size = 3))
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function 
```

```python
x = np.linspace(-5, 5, 100)

y = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))


data = {

    'X': x.flatten(),

    'Y': y.flatten(),

    'Z': z.flatten()

}

df = pd.DataFrame(data)
```

```python
df
```

```python
px.scatter_3d(x = df.X, y = df["Y"], z= df.Z).show()
```
